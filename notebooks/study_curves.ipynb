```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
list_long_cycle = [2,5,10,20]
```

```python
bidule = np.array([2,3,4])
```

```python
original_acc = []
mixed_same_acc = []
mixed_rand_acc = []
for n_epoch_cycle in list_long_cycle :
    original_acc.append(np.load("../curves/original_acc_" + str(n_epoch_cycle) + ".npy"))
    mixed_same_acc.append(np.load("../curves/mixed_same_acc_" + str(n_epoch_cycle) + ".npy"))
    mixed_rand_acc.append(np.load("../curves/mixed_rand_acc_" + str(n_epoch_cycle) + ".npy"))
```

```python
for courbe, n_epoch_cycle in zip(mixed_same_acc, list_long_cycle) :
    plt.plot(range(len(courbe)), courbe, label = "n_epoch_cycle " + str(n_epoch_cycle))

plt.legend()
plt.show
```

```python
for courbe, n_epoch_cycle in zip(mixed_rand_acc, list_long_cycle) :
    plt.plot(range(len(courbe)), courbe, label = "n_epoch_cycle " + str(n_epoch_cycle))

plt.legend()
plt.show
```
