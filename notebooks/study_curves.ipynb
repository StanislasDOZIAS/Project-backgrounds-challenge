```python
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
list_long_cycle = [2,5,10,15,20]
```

```python
bidule = np.array([2,3,4])
```

```python
original_acc = []
mixed_same_acc = []
mixed_rand_acc = []
for n_epoch_cycle in list_long_cycle :
    original_acc.append(np.load("../curves/original_acc_" + str(n_epoch_cycle) + ".npy"))
    mixed_same_acc.append(np.load("../curves/mixed_same_acc_" + str(n_epoch_cycle) + ".npy"))
    mixed_rand_acc.append(np.load("../curves/mixed_rand_acc_" + str(n_epoch_cycle) + ".npy"))
```

```python
for i, (courbe, n_epoch_cycle) in enumerate(zip(mixed_same_acc, list_long_cycle)) :
    plt.plot(range(len(courbe)), courbe, label = "n_epoch_cycle " + str(n_epoch_cycle), color=cm.cool(i/len(mixed_same_acc)))

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show
```

```python
for i, (courbe, n_epoch_cycle) in enumerate(zip(mixed_rand_acc, list_long_cycle)) :
    plt.plot(range(len(courbe)), courbe, label = "n_epoch_cycle " + str(n_epoch_cycle), color=cm.cool(i/len(mixed_same_acc)))

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show
```
